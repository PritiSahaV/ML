# **Loading Dataset:**

```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
wine_data = pd.read_csv('/content/sample_data/wine.csv')
wine_data.head(10)
```

# **Encoding Categorical Features:**

```python
wine_data.info()
```

*THE "QUALITY" FEATURE HAS OBJECT TYPE DATA.*

```python
wine_data['quality'].unique()
```

*'QUALITY' IS A CATEGORICAL FEATURE AS IT IS DIVIDED INTO TWO CATEGORIES, "BAD" AND "GOOD". SO, WE CAN USE BINARY ENCODING FOR THIS FEATURE.*

```python
from sklearn.preprocessing import LabelEncoder

# Set up the LabelEncoder object
enc = LabelEncoder()

# Apply the encoding to the "quality" column
wine_data['quality_enc'] = enc.fit_transform(wine_data['quality'])

# Compare the two columns
print(wine_data[['quality', 'quality_enc' ]].head(10))
```

```python
#DROPPING THE "QUALITY" COLUMN
wine_data = wine_data.drop(['quality'], axis = 1)
wine_data.head(10)
```

# **Feature Selection:**

Checking for correlated features.

```python
wine_corr = wine_data.corr()
wine_corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
sns.heatmap(wine_corr, cmap = 'YlGnBu', annot=True)
```

```python
#DROPPING THE LESS CORRELATED COLUMNS('pH', 'residual sugar') WITH LABEL quantity_enc AND 'free sulfur dioxide' FEATURE AS IT IS HIGHLY CORRELATED WITH 'total sulfur dioxide'
wine_data = wine_data.drop(['pH', 'residual sugar', 'free sulfur dioxide'], axis = 1)
wine_data.head(5)
```

# **Splitting Dataset into Train and Test:**

Spliting the dataset into features and labels.

```python
#Prepare the training set

# x = feature values, all the columns except the last column
x = wine_data.iloc[:, :-1]

# y = target values, last column of the data frame
y = wine_data.iloc[:, -1]
```

```python
print("Feature data:\n",x)
print("Target data:\n",y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42, stratify = y)
```

```python
print("x train shape:",x_train.shape)
print("x test shape:",x_test.shape)
print("y train shape:",y_train.shape)
print("y test shape:",y_test.shape)
```

# **Scaling Dataset using MinMax Scaler:**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(x_train)
```

```python
# transform data
x_train_scaled = scaler.transform(x_train)
```

```python
print("per-feature minimum before scaling:\n {}".format(x_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(x_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(x_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(x_train_scaled.max(axis=0)))
```

```python
# transform test data
x_test_scaled = scaler.transform(x_test)
```

# **Logistic Regression:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(x_train_scaled, y_train) #Training the model
predictions = model.predict(x_test_scaled)
print(predictions)# printing predictions
```

```python
Logistic_score= accuracy_score(y_test, predictions)
print(Logistic_score)
```

# **Decision Tree:**

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy',random_state=1)
clf.fit(x_train_scaled,y_train)
y_pred = clf.predict(x_test_scaled)
Decision_Tree_score=accuracy_score(y_pred,y_test)
print(Decision_Tree_score)
```

# **Plot:**

```python
plot_data= pd.DataFrame({"Models": ['Logistic Regression', 'Decision Tree'], "Score":[Logistic_score, Decision_Tree_score]})
plt.figure(figsize = (10, 8))

sns.barplot(x = 'Models', y = 'Score', data = plot_data, palette="Blues_d")
plt.ylim(0, 0.85)
plt.show()
```
